```python
#extension = 'fra+chi_sim'
extension = 'pf1_r'
#l = 'chi_sim+fra'
print(pytesseract.image_to_data(Image.open(filename), lang = l))
# ISSUE:The printing result and the write-in result 
f = open("output/all-{}.txt".format(extension), 'w')
string = pytesseract.image_to_data(Image.open(filename), lang = l)
f.write(string)
f.close()
```

```python
f = open("output/boxes-{}.txt".format(extension), 'w')
f.write(pytesseract.image_to_boxes(Image.open(filename), lang = l))
f.close()
```

```python
# <symbol/tab> <left> <bottom> <right> <top> <page>
with open('output/boxes-{}.txt'.format(extension), 'r') as f:
    lines = f.readlines()
    boxes = []
    for line in lines:
        boxes.append(line.strip('\n').split(' '))
    #print(boxes)
    f.close()
```

```python
# 'level', 'page_num', 'block_num', 'par_num', 'line_num', 'word_num', 'left', 'top', 'width', 'height', 'conf', 'text'
with open('output/all-{}.txt'.format(extension), 'r') as f:
    lines = f.readlines()
    #print(lines)
    lo = []
    for line in lines[1:]:
        lo.append(line.strip('\n').split('\t'))
    f.close()
#lo
```

```python
# count for each char
im = Image.open(filename)
width, height = im.size

import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
img = plt.imread(filename)
img = np.flipud(img)
fig, ax = plt.subplots()
ax.set_ylim([0, height])
ax.imshow(img)

x = []
y = []
for i in boxes:
    x.append(float(i[1]))
    y.append(float(i[2]))

from time import sleep
for num in range(len(x)):
    ax.add_patch(
    
        patches.Rectangle(
            (x[num],y[num]),
            float(boxes[num][3]),
            float(boxes[num][4]),
            fill = False
        )
    )
    #sleep(0.1)
len(x)
#ax.plot(x,y)
```
