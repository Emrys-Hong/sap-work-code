# PCA 
## by converting the BnW image to 2D points

```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
import cv2
```

```python
im = cv2.imread('/Users/i351707/Desktop/cropped_images2text/1416.png', 0) # change the image path
ret,thresh_img = cv2.threshold(im,120,255,cv2.THRESH_BINARY)
plt.imshow(thresh_img[:], cmap='gray')
```

```python
#thresh_img[10]
```

```python
import pandas as pd
def cvtImageToPoints(binImg):
    Xs = []
    Ys = []
    data = []
    h, w = binImg.shape
    print(h,w)
    for i in range(h):
        for j in range(w):
            if binImg[i][j] == 0:
                Xs.append(j)
                Ys.append(i)
                data.append([j,i])
    return np.asarray(data), np.asarray(Xs), np.asarray(Ys)
    
```

```python
data, X, Y = cvtImageToPoints(thresh_img)
#plt.ylim(thresh_img.shape[0], 0)
#plt.plot(Y,X, '.')
```

```python
# Find the empirical mean along each column
mean_vec = np.mean(data, axis=0)
print(mean_vec)
print(max(X), max(Y))
# Calculate the deviations from the mean
cov_mat = (data - mean_vec).T.dot((data - mean_vec)) / (data.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
cov_mat = np.cov(data.T)
#print(cov_mat)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)

eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
eig_pairs.sort(key=lambda x: x[0], reverse=True)

fig, ax = plt.subplots()
plt.ylim(thresh_img.shape[0], 0)
ax.scatter(X, Y, marker='.', c='black')
ax.quiver([thresh_img.shape[0]/2,thresh_img.shape[0]/2], [thresh_img.shape[0]/2,thresh_img.shape[0]/2], eig_pairs[0][1], eig_pairs[1][1], scale=10, color=['r','g'])
plt.show()
```

```python
# The direction of the red component (the first component) 
# suggests the writing sequence.
first_cpts = [eig_pairs[0][1][0], eig_pairs[1][1][0]]
print(first_cpts)
```
