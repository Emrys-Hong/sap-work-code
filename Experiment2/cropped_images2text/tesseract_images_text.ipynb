```python
def skew_angle(img):
    # add to see whether the image is correctly horizontal
    horizontal = True if img.shape[0]< img.shape[1] else False
    
    img = cv2.threshold(img, 0, 255,
	cv2.THRESH_BINARY | cv2.THRESH_OTSU)[1]
    
    edges = cv2.Canny(img,50,150,apertureSize = 3)
    
#     cv2.imwrite('result.png', edges)
    lines = cv2.HoughLines(edges,1,np.pi/180,50)

    angle_sum=0;count=0
    for coordinates in lines:
        for rho,theta in coordinates:
            a = np.cos(theta)
            b = np.sin(theta)
            x0 = a*rho
            y0 = b*rho
            x1 = int(x0 + 1000*(-b))
            y1 = int(y0 + 1000*(a))
            x2 = int(x0 - 1000*(-b))
            y2 = int(y0 - 1000*(a))

            
            angle = math.atan2((y2-y1),(x2-x1))/math.pi*180
            if  horizontal:
                if angle < 45:
                    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
                    angle_sum += angle
                    count += 1
            else:
                if angle > 45:
                    cv2.line(img,(x1,y1),(x2,y2),(0,0,255),2)
                    angle_sum += angle
                    count += 1
    angle_ave = angle_sum / count
    return angle_ave 

def rotate_vertical(image, angle):
    # grab the dimensions of the image and then determine the
    # center
    (h, w) = image.shape[:2]
    (cX, cY) = (w / 2, h / 2)
    
    # grab the rotation matrix (applying the negative of the
    # angle to rotate clockwise), then grab the sine and cosine
    # (i.e., the rotation components of the matrix)
    M = cv2.getRotationMatrix2D((cX, cY), -angle, 1.0)
    cos = np.abs(M[0, 0])
    sin = np.abs(M[0, 1])

    # compute the new bounding dimensions of the image
    nW = int((h * sin) + (w * cos))
    nH = int((h * cos) + (w * sin))

    # adjust the rotation matrix to take into account translation
    M[0, 2] += (nW / 2) - cX
    M[1, 2] += (nH / 2) - cY

    # perform the actual rotation and return the image
    return cv2.warpAffine(image, M, (nW, nH),borderMode=0, borderValue=(255,255,255))


def deskew_horizontal(image, angle):
    (h, w) = image.shape[:2]
    center = (w // 2, h // 2)
    M = cv2.getRotationMatrix2D(center, angle, 1.0)
    rotated = cv2.warpAffine(image, M, (w, h),
                             flags=cv2.INTER_CUBIC, borderMode=cv2.BORDER_REPLICATE)
    return rotated

def img2txt(address):
    img = cv2.imread(address, cv2.IMREAD_GRAYSCALE)
    try:
        angle = skew_angle(img)
        if angle < 45:
            deskewed_image = deskew_horizontal(img.copy(),angle)
            cv2.imwrite('/Users/i351707/Desktop/result.png',deskewed_image)
        elif angle >= 45:
            deskewed_image = deskew_vertical(img.copy(),angle)
            cv2.imwrite('/Users/i351707/Desktop/result.png',deskewed_image)
        txt = image_to_string(deskewed_image,lang='tha+eng')
        print(True)
    except Exception:
        try:
            txt = image_to_string(img,lang='tha+eng')
        except Exception as e:
            txt = e.args
    return txt

```
